# Sampling the data #

```python
#code by gemini 3
import pandas as pd
import os

input_path = r"C:\Users\RoyB\Downloads\Motor_Vehicle_Collisions_-_Crashes.csv"
# שיניתי את שם הקובץ ביציאה כי כעת הוא מכיל גם הולכי רגל
output_path = r"C:\Users\RoyB\Downloads\VRU_Crashes_Sample.csv"

# הגדרת גודל המנה
chunk_size = 100000 
first_chunk = True
total_rows_saved = 0

print("Starting processing with Pedestrians and Cyclists...")

# קריאת הקובץ בחלקים
with pd.read_csv(input_path, chunksize=chunk_size, low_memory=False) as reader:
    for i, chunk in enumerate(reader):
        
        # בניית המסכה (התנאי) הלוגית המורחבת
        # נשמור שורה אם: (אופניים נפצעו) או (אופניים נהרגו) או (הולכי רגל נפצעו) או (הולכי רגל נהרגו)
        condition = (
            (chunk['NUMBER OF CYCLIST INJURED'] > 0) | 
            (chunk['NUMBER OF CYCLIST KILLED'] > 0) |
            (chunk['NUMBER OF PEDESTRIANS INJURED'] > 0) | 
            (chunk['NUMBER OF PEDESTRIANS KILLED'] > 0)
        )
        
        filtered_chunk = chunk[condition]
        
        # שמירת הנתונים אם יש שורות רלוונטיות במנה
        if not filtered_chunk.empty:
            if first_chunk:
                filtered_chunk.to_csv(output_path, index=False, mode='w')
                first_chunk = False
            else:
                filtered_chunk.to_csv(output_path, index=False, mode='a', header=False)
            
            total_rows_saved += len(filtered_chunk)
        
        # הדפסת התקדמות כל 5 מנות
        if i % 5 == 0:
            print(f"Processed chunk {i+1}...")

print("------------------------------------------------")
print(f"Process Complete!")
print(f"Total rows saved (Cyclists & Pedestrians): {total_rows_saved}")
print(f"File saved at: {output_path}")
```
